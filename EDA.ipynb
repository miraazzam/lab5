```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cleaned_ebay_deals.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])

df = df.sort_values("timestamp")

df["hour"] = df["timestamp"].dt.hour

deals_per_hour = df.groupby("hour").size()

plt.figure()
deals_per_hour.plot(kind="bar")
plt.title("Number of Deals per Hour")
plt.xlabel("Hour")
plt.ylabel("Number of Deals")
plt.show()
```

```python
plt.figure()
plt.hist(df["price"].dropna())
plt.title("Price Distribution")
plt.show()

plt.figure()
sns.boxplot(x=df["price"])
plt.title("Price Boxplot")
plt.show()
```

```python
plt.figure()
plt.scatter(df["original_price"], df["price"])
plt.xlabel("Original Price")
plt.ylabel("Discounted Price")
plt.title("Original vs Discounted Price")
plt.show()
```

```python
plt.figure()
plt.hist(df["discount_percentage"].dropna())
plt.title("Discount Percentage Distribution")
plt.show()
```

```python
shipping_counts = df["shipping"].value_counts()

plt.figure()
shipping_counts.plot(kind="bar")
plt.title("Shipping Options Frequency")
plt.show()
```

```python
keywords = ["Apple","Samsung","Laptop","iPhone","Tablet","Gimbal"]

counts = {}

for word in keywords:
    counts[word] = df["title"].str.contains(word, case=False, na=False).sum()

plt.figure()
plt.bar(counts.keys(), counts.values())
plt.title("Keyword Frequency in Titles")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Define keywords
keywords = ["Apple", "Samsung", "Laptop", "iPhone", "Tablet", "Gimbal"]
counts = {}

# 2. Count occurrences in the 'title' column
for word in keywords:
    counts[word] = df["title"].str.contains(word, case=False, na=False).sum()

# 3. Create the Bar Chart
plt.figure(figsize=(10, 6))
plt.bar(counts.keys(), counts.values(), color='skyblue', edgecolor='navy')

# 4. Add labels and title
plt.xlabel("Product Keywords")
plt.ylabel("Number of Items Found")
plt.title("Keyword Frequency in eBay Tech Deals")

# 5. Show the plot
plt.show()

```

```python
df["price_difference"] = df["original_price"] - df["price"]

plt.figure()
plt.hist(df["price_difference"].dropna())
plt.title("Price Difference Distribution")
plt.show()
```
